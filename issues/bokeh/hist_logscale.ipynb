# Bokeh log-axis plot

When using a code I had written some time ago, for Python-2/Bokeh-0.11*ish*, I noticed a blank plot.
No errors (after Python-3 adjustments), but a plot with *no* data shown; there were axes, legend, tool, but no `glyphs` in it.
Now I was using Bokeh `0.12.6`.

After all tests I could imagine regarding the data content itself, I tried the axis scale: I changed from `log` to `linear` and my data was back.
So, here, I'll (try to) reproduce and then report --if-- the bug.

```python
import sys; print('Python version:{}'.format(sys.version.split()[0]))
import bokeh; print('Bokeh version:{}'.format(bokeh.__version__))
import numpy; print('Numpy version:{}'.format(numpy.__version__))
```

```python
from bokeh.io import output_notebook
output_notebook()
```

```python
def plot_hist(data,yscale):
    from bokeh.plotting import figure
    from numpy import histogram

    h,b = histogram(data)
    
    tops = h
    bottoms = [0] * len(h)
    lefts = b[:-1]
    rights = b[1:]

    color = '#FFCC00'
    alpha = 1

    fig = figure(y_axis_type=yscale)

    fig.quad(top=tops,
            bottom=bottoms,
            left=lefts,
            right=rights,
            fill_color=color,fill_alpha=alpha,
            legend='hist-test')
    
    return fig
```

```python
from bokeh.io import show

data = numpy.random.rand(1000)

plt = plot_hist(data,'log')

show(plt)
```

```python
from bokeh.io import show
plt = plot_hist(data,'linear')
show(plt)
```
